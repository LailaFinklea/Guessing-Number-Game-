```python
import random # Import the random module to generate the secret number

def get_range():
    print("Welcome To Our Guess the Number Game!")
    try:
        # Ask user for the lower and upper bound of the guessing range
        lower_bound = int(input("Enter the LOWER bound of your range: "))
        upper_bound = int(input("Enter the HIGHER bound of your range: "))
        
        # Ensure the lower bound is lower than the upper bound
        if lower_bound >= upper_bound:
            x = False 
            while x == False: # Loop until a valid range is provided
                print("Lower bound MUST be less than the upper bound! Please try again.")
                lower_bound = int(input("Enter the LOWER bound of your range: "))
                upper_bound = int(input("Enter the HIGHER bound of your range: "))
                if lower_bound < upper_bound:
                    x = True # Break the loop once bounds are correct
    
        # Select a random number within the user's defined range
        secret_numb = random.randint(lower_bound, upper_bound)
        return secret_numb, lower_bound, upper_bound # Return game data as a tuple
    
    except ValueError:
        # Ensure the user inputs an integer value
        print("Invalid input. Please enter whole numbers for the bounds")
        return None


def get_user_guess(lower_bound, upper_bound):
    while True: # Keep asking until a valid guess is made
        try:
            guess = int(input(f"Please enter your guess ({lower_bound}-{upper_bound}): "))
            # Check if the guess is actually inside the range
            if lower_bound <= guess <= upper_bound:
                return guess # Valid guess returned
            else:
                print("Invalid input: Please guess within your range")
        except ValueError:
            # Catch errors if the guess isn't an integer
            print("Invalid input: Please enter an integer")


def compare_guess(guess, secret_numb):
    # Check if the guess is higher than the secret number
    if guess > secret_numb:
        print("Too high!") 
        return "low"
    # Check if the guess is lower than the secret number
    if guess < secret_numb:
        print("Too low!") 
        return "high"
    # Check if the guess is correct
    if guess == secret_numb: 
        print("Correct!")
        return "win"  


is_running = True # Control variable for the main game loop

while is_running == True:
    range_data = get_range() # Call function to set up the game
    
    if range_data != None: # Only proceed if the range was set up correctly
        secret = range_data[0] # The target number
        low = range_data[1] # The lower limit
        high = range_data[2] # The upper limit
        
        attempts = 0 # Counter for how many guesses the user takes
        previous_guesses = [] # List to track guess history
        status = "" # Variable to track if the user has won
        
        # Keep playing the round until the compare_guess returns "win"
        while status != "win":
            user_guess = get_user_guess(low, high)
            attempts = attempts + 1 
            
            previous_guesses.append(user_guess) # Save guess to history
            print(f"Previous guesses: {previous_guesses}")
            
            status = compare_guess(user_guess, secret) # Update status
        
        print(f"It took you {attempts} attempts!")        
        
        # Save the number of attempts to an external file
        with open("highscore.txt", "w") as file:
            file.write(str(attempts))

    # Ask the user if they want to restart or quit
    play_again = input("Do you want to play again? (yes/no): ")
    if play_again != "yes":
        is_running = False # Exit the main loop
        print("Goodbye!")
```

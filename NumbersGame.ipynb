## Guess the Number Game!

```python
import random

is_running = True

def get_range (lower_bound,upper_bound):
    print("Welcome To Our Guess the Number Game!")
    try:
        lower_bound = int(input("Enter the LOWER bound of your range: "))
        upper_bound = int(input("Enter the HIGHER bound of your range:  "))
        
        if lower_bound>= upper_bound:
            print ("Lower bound MUST be less then the upper bound! Please try again. ")
            
        else: 
            secret_numb = random.randint (lower_bound,upper_bound)
            return secret_numb
        return lower_bound and upper_bound
    except ValueError:
        print ("Invaild input. please enter whole numbers for the bounds")


def get_user_guess(lower_bound,upper_bound):
    while True:
        try:
            guess = int(input("Please enter your guess within range"))
            if lower_bound <= guess <= upper_bound:
                return guess
            else:
                print("invalid input: please guess within your range")
        except ValueError:
            print("Invalid input: please enter an integer")


def compare_guess(guess,secret_numb):
    if guess > secret_numb:
        print("Your guess is high") 
    if guess < secret_numb:
        print("Your guess is low") 
    if guess == secret_numb: 
        print("Winner!")
        return "win"
```

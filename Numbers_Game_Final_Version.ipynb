```python
import random

def get_range():
    print("Welcome To Our Guess the Number Game!")
    try:
        lower_bound = int(input("Enter the LOWER bound of your range: "))
        upper_bound = int(input("Enter the HIGHER bound of your range: "))
        
        if lower_bound >= upper_bound:
            x = False 
            while x == False:
                print("Lower bound MUST be less than the upper bound! Please try again.")
                lower_bound = int(input("Enter the LOWER bound of your range: "))
                upper_bound = int(input("Enter the HIGHER bound of your range: "))
                if lower_bound < upper_bound:
                    x = True
    
        secret_numb = random.randint(lower_bound, upper_bound)
        return secret_numb, lower_bound, upper_bound
    
    except ValueError:
        print("Invalid input. Please enter whole numbers for the bounds")
        return None


def get_user_guess(lower_bound, upper_bound):
    while True:
        try:
            guess = int(input(f"Please enter your guess ({lower_bound}-{upper_bound}): "))
            if lower_bound <= guess <= upper_bound:
                return guess
            else:
                print("Invalid input: Please guess within your range")
        except ValueError:
            print("Invalid input: Please enter an integer")


def compare_guess(guess, secret_numb):
    if guess > secret_numb:
        print("Too high!") 
        return "low"
    if guess < secret_numb:
        print("Too low!") 
        return "high"
    if guess == secret_numb: 
        print("Correct!")
        return "win"  


is_running = True 

while is_running == True:
    range_data = get_range()
    
    if range_data != None:
        secret = range_data[0]
        low = range_data[1]
        high = range_data[2]
        
        attempts = 0 
        previous_guesses = [] 
        status = ""
        
        while status != "win":
            user_guess = get_user_guess(low, high)
            attempts = attempts + 1
            
            previous_guesses.append(user_guess)
            print(f"Previous guesses: {previous_guesses}")
            
            status = compare_guess(user_guess, secret)
        
        print(f"It took you {attempts} attempts!")        
        
        with open("highscore.txt", "w") as file:
            file.write(str(attempts))

    play_again = input("Do you want to play again? (yes/no): ")
    if play_again != "yes":
        is_running = False
        print("Goodbye!")
```
